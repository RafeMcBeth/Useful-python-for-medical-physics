# First load and plot profile data

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
raw = pd.read_csv("6MV 100FSD PROFILE (PDD JUNE - RAW).csv") # , index_col=[0]
```

```python
raw.loc[:10]   # first 10
```

```python
field = raw.loc[3].tolist()  # get depth row and use as column names
#field
```

```python
depth = raw.loc[4].tolist()  # get depth row and use as column names
#depth
```

```python
col_names = [a + ' : ' + b for a,b in zip(field,depth)]  # drop 'Fieldsize: : Depth:',
col_title = col_names.pop(0)
col_names
```

```python
Profile = raw.ix[7:]    # get PDD data from row 7 onwards
Profile.columns = depth  # set the column names
Profile = Profile.set_index('Depth:')  # set index
Profile.columns = col_names  # set the column names
Profile = Profile.fillna(0).astype(float)  # set NaN to 0 and return floats
Profile.head()
```

```python
Profile.plot().legend(loc='center left', bbox_to_anchor=(1, 0.5), title=col_title); # label=field  legend=False
```

# Plot PDD data

```python
PDDraw = pd.read_csv("LA6(2015) 6MV 100FSD PDD - RAW.csv") # , index_col=[0]
```

```python
PDDraw[:10]
```

```python
PDDfield = PDDraw.loc[3].tolist()  # get depth row and use as column names
PDDdepth = PDDraw.loc[4].tolist()  # get depth row and use as column names
PDDcol_names = [a + ' : ' + b for a,b in zip(PDDfield,PDDdepth)]  # drop 'Fieldsize: : Depth:',
PDDcol_title = PDDcol_names.pop(0)
#PDDcol_names
```

```python
PDD = PDDraw.ix[7:]    # get PDD data from row 7 onwards
PDD.columns = PDDdepth  # set the column names
PDD = PDD.set_index('Depth:')  # set index
PDD.columns = PDDcol_names  # set the column names
PDD = PDD.fillna(0).astype(float)  # set NaN to 0 and return floats
PDD.head()
```

```python
PDD.plot().legend(loc='center left', bbox_to_anchor=(1, 0.5), title=PDDcol_title); # label=field  legend=False
```
