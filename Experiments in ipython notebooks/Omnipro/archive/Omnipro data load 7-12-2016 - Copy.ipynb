# First load and plot profile data

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
profileRaw = pd.read_csv("6MV 100FSD PROFILE (PDD JUNE - RAW).csv") # , index_col=[0]
```

## Get the info
0 column is info, 1: are the data, each col being another spectra

```python
profile_info_df = profileRaw.loc[:4]
#profile_info_df
```

Get the info strings and clean

```python
description = profile_info_df['0'].tolist()
#print(description)
description2 = [description[i].split(":")[0].replace(" ", "") for i in range(len(description))]  # remove : and strip whitespaces
description2
```

```python
profile_info_df = profile_info_df.drop('0', 1)   # drop the now refunct 0 column
```

```python
ID_list = [':'.join(profile_info_df[i].tolist()) for i in list(profile_info_df.columns)]  # iterate over columns, joining entries by : and creating a list
#ID_list
```

```python
profile_info_df = profile_info_df.transpose()  # access methods require selection by keys which should be columns, therefore transpose
profile_info_df.columns = description2
profile_info_df['Description'] = ID_list
profile_info_df
```

## Get the numerial profile data

```python
profile_data_df = profileRaw.ix[7:]    # get numerical data from row 7 onwards
profile_data_df=profile_data_df.rename(columns = {'0':'Depth'}).set_index('Depth')  # rename the column and set as index
```

```python
profile_data_df.columns = ID_list  # set the column names from the ID_list
profile_data_df = profile_data_df.fillna(0).astype(float)  # set NaN to 0 and return floats
profile_data_df.head()
```

```python
# profile_data_df.plot().legend(loc='center left', bbox_to_anchor=(1, 0.5)); # label=field  legend=False, title=col_title
```

## Filter and plot

```python
to_plot = profile_info_df['Description'][(profile_info_df['Depth'] == '100.0 mm') & (profile_info_df['ScanType'] == 'Crossline') ].tolist()  # get description field where...
to_plot
```

```python
profile_data_df[to_plot].plot().legend(loc='center left', bbox_to_anchor=(1, 0.5));
```
