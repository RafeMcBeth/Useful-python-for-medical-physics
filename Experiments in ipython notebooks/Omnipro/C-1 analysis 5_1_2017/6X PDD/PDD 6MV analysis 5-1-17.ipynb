```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
from scipy import interpolate
```

```python
os.listdir('data/')  
```

```python
CC13 = pd.read_csv('data/CC13 100FSD PDDs - 100x100 only.csv', index_col = 0)
diamond  = pd.read_csv('data/Micro diamond 6MV PDD 100x100 only.csv', index_col = 0)
Diode  = pd.read_csv('data/Photon Diode 6MV PDDs 100x100 only.csv', index_col = 0)
```

```python
fig_width = 12
fig_height = 6
plt.figure(figsize=(fig_width,fig_height))

plt.plot(CC13, color='b', label='CC13')
plt.plot(diamond, color='g', label='Micro diamond')
plt.plot(Diode, color='r', label='Diode')

plt.xlim(-10,100)
plt.ylim(50, 125)

plt.title('6MV PDD for 100 x 100 field, 100FSD, three detectors')
plt.ylabel('PDD (%)')
plt.xlabel('Distance (mm)')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5));
```
