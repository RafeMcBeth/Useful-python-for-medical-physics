```python
import os
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
files = os.listdir() 
```

```python
files[1]
```

```python
files[2]
```

```python
AAA_df = pd.read_table(files[1], skiprows=11, index_col=0, header=None, names=['PDD'])
AAA_df = 100.0*AAA_df/AAA_df.max()   # normalise

AXB_df = pd.read_table(files[2], skiprows=11, index_col=0, header=None, names=['PDD'])
AXB_df = 100.0*AXB_df/AXB_df.max()   # normalise
```

```python
files[5]
```

```python
measured_df = pd.read_csv(files[5], index_col=0)    # the 6MV LA6 2015 data
measured_df = 100.0*measured_df/measured_df.max()   # normalise
measured_df.index = measured_df.index.values/10.0   # make cm
# measured_df.plot()
```

```python
height=10
width=7           # wwidth of figs
f, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(width, height))

ax1.plot(AAA_df, label = 'AAA', linestyle='--', c='r', marker='x')
ax1.plot(AXB_df, label = 'AXB', linestyle='--', c='b', marker='x')
ax1.plot(measured_df, label = 'Measured',  c='g', marker='x')
ax1.legend(loc=4)
ax1.set_xlabel('Depth (cm)')
ax1.set_ylabel('PDD (%)')
#ax1.set_xlim([20,25])
ax1.set_xlim([0,0.5])
#x1.set_ylim([27,37])

ax2.plot(AAA_df, label = 'AAA', linestyle='--', c='r')   #, marker='x'
ax2.plot(AXB_df, label = 'AXB', linestyle='--', c='b')
ax2.plot(measured_df, label = 'Measured',  c='g')
ax2.legend(loc=1)
ax2.set_xlabel('Depth (cm)')
ax2.set_ylabel('PDD (%)')
ax2.set_xlim([20,25])
#ax2.set_xlim([0,0.5])
ax2.set_ylim([27,37])


ax3.plot(AAA_df-AXB_df, label = 'AAA - AXB', c='m')
ax3.set_ylim([-2,2])
ax3.set_xlabel('Depth (cm)')
ax3.set_ylabel('PDD difference AAA - AXB (%)')
ax3.legend(loc=4)

plt.show()
```
