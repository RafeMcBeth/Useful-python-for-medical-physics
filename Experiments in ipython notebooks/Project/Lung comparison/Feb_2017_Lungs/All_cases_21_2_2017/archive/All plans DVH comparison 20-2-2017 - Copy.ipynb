## Do wilcox test on paired DVH

```python
case_list = AXB_dict.keys()
case_list = [entry.strip('_AXB') for entry in case_list]
case_list
```

```python
# loop through the DVH annd create two dataframes with matching columns 

for i, key in enumerate(case_list):
    temp_AAA_DVH = AAA_dict.get(key + '_AAA').DVH_df[structure]   # get the DVH data
    temp_AAA_DVH_df = pd.DataFrame({key : temp_AAA_DVH})       # place in a df
    
    temp_AXB_DVH = AXB_dict.get(key + '_AXB').DVH_df[structure]   # get the DVH data
    temp_AXB_DVH_df = pd.DataFrame({key : temp_AXB_DVH})       # place in a df

    if i == 0:
        AAA_df_wilcox = temp_AAA_DVH_df  # create the dataframe
        AXB_df_wilcox  = temp_AXB_DVH_df  # create the dataframe
    else:       
        AAA_df_wilcox  = pd.concat([AAA_df_wilcox , temp_AAA_DVH_df], axis=1)   # if df exists, populate    
        AXB_df_wilcox  = pd.concat([AXB_df_wilcox , temp_AXB_DVH_df], axis=1)   # if df exists, populate
        
    AAA_df_wilcox  = AAA_df_wilcox.fillna(value=0)
    AXB_df_wilcox  = AXB_df_wilcox.fillna(value=0)
```

```python
index = AAA_df_wilcox.index.values[-1]
temp_df  = pd.concat([AAA_df_wilcox.loc[index], AXB_df_wilcox.loc[index]], axis=1)  
temp_df.columns = ['AAA', 'AXB']
temp_df
```

```python
my_wilcox(temp_df.diff(axis = 1).loc[:, 'AXB'].values)   # get p of the differences
```

```python
def get_my_wilcox(index):    # convenience function with dataframes hard coded
    temp_df  = pd.concat([AAA_df_wilcox.loc[index], AXB_df_wilcox.loc[index]], axis=1)  
    temp_df.columns = ['AAA', 'AXB']
    z_stat, p_val = wilcoxon(temp_df.diff(axis = 1).loc[:, 'AXB'].values)   # get p of the differences
    if np.isnan(p_val):  # if no difference (i.e. both have 100%) then return 1
        return 1.0
    else:
        return p_val
```

```python
wilcox_results = [get_my_wilcox(index) for index in AAA_df_wilcox.index.values]    # for each dose index get the wilcox 
wilcox_results = pd.Series(wilcox_results, index=AAA_df_wilcox.index.values)
```

Plot the wilcox significance at each dose level

```python
width=8
height=2
plt.figure(figsize=(width, height))


plt.plot(wilcox_results)
plt.axhline(y=0.05, c='k', ls = '--')   # add line to show significant

plt.fill_between(wilcox_results[wilcox_results < 0.05].index.values, wilcox_results[wilcox_results < 0.05].values,y2=0.05, color = 'r')
plt.xlim([50,80])
plt.title(structure + ' Wilcoxon significance of dose differences')
plt.xlabel('Dose (Gy)')
plt.ylabel('P value')
```
