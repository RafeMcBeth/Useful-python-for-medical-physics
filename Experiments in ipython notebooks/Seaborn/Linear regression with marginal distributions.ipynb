```python
import seaborn as sns
sns.set(style="darkgrid", color_codes=True)
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
g = sns.jointplot("total_bill", "tip", data=tips, kind="reg",
                  xlim=(0, 60), ylim=(0, 12), color="r", size=7)
```
