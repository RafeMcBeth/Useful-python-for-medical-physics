```python
import requests
import json
import pprint
```

```python
url = 'http://environment.data.gov.uk/flood-monitoring/id/floods'
r = requests.get(url)  # get the response object
```

```python
def print_json(json_data):
    pprint.PrettyPrinter().pprint(json_data)
    
print_json(r.json()['items'])
```

```python
floods = [item['description'] for item in r.json()['items']]
```
