```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import statsmodels.api as sm
from statsmodels import regression
```

```python
x = np.random.normal(0.01,1,1000)
b = np.random.normal(0.05,1,1000)
y = np.cumsum(x)
c = np.cumsum(b)
d = y+2*c
plt.plot(y)
plt.plot(c)
plt.plot(d, 'ro')
```

```python
slrc = regression.linear_model.OLS(d, sm.add_constant(c))
slry = regression.linear_model.OLS(d, sm.add_constant(y))
```

```python
results = slrc.fit()
results.summary()
```

```python
results = slry.fit()
results.summary()
```
