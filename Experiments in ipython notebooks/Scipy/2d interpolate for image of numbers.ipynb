## Read in array of data from csv file, plot as image then test interpolation 

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import interpolate
```

http://docs.scipy.org/doc/numpy/reference/generated/numpy.loadtxt.html#numpy.loadtxt

```python
data = pd.read_csv("test.csv",  index_col=[0])  # make sure numerical data is type 'number' in excel to avoid problems
data
```

```python
data_ = data.as_matrix()  # get the numerical data
print(data_.shape)  # check size of data
plt.imshow(data_);  # show the raw data
```

from http://matplotlib.org/examples/images_contours_and_fields/interpolation_methods.html

```python
methods = [None, 'none', 'nearest', 'bilinear', 'bicubic', 'spline16',
           'spline36', 'hanning', 'hamming', 'hermite', 'kaiser', 'quadric',
           'catrom', 'gaussian', 'bessel', 'mitchell', 'sinc', 'lanczos']


fig, axes = plt.subplots(3, 6, figsize=(12, 6),
                         subplot_kw={'xticks': [], 'yticks': []})

fig.subplots_adjust(hspace=0.3, wspace=0.05)

for ax, interp_method in zip(axes.flat, methods):
    ax.imshow(data_, interpolation=interp_method)
    ax.set_title(interp_method)

plt.show()
```
